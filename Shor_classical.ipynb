```python
import math
math.gcd(a, N)
```

```python
import matplotlib.pyplot as plt
z = list(range(N))
y = [a**z0 % N for z0 in z]

plt.plot(z, y)
plt.xlabel('z')
plt.ylabel('{}^z mod{}'.format(a, N))
plt.show()
```

```python
r = z[y[1:].index(1)+1]
print(f'r = {r}')
```

```python
if r % 2 == 0:
    x = (a**(r/2.)) % N
    print(f'x = {x}')
    if ((x + 1) % N) != 0:
        print(math.gcd((int(x)+1), N), math.gcd((int(x)-1), N))
    else:
        print("x + 1 is 0 (mod N)")
else:
    print('r + {} is odd'.format(r))
```
